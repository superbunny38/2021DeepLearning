## 1.4.2 클래스

```python
class Man:
    def __init__(self,name):#생성자: 클래스를 초기화하는 방법
        self.name = name
        print("Initialized!")
        
    def hello(self):
        print("Hello"+self.name+"!")
        
    def goodbye(self):
        print("Good-bye "+self.name+"!")
m = Man("David")#객체 생성
m.hello()
m.goodbye()
```

## 1.6 matplotlib
- 데이터 그리기와 데이터 시각화

```python
import matplotlib.pyplot as plt
```

- sin 함수 그리기

```python
x = np.arange(-5,5,0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

## 1.6.2 pyplot의 기능

```python
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
```

```python
plt.figure(figsize=(15, 7))
plt.plot(x,y1,label="sin")
plt.plot(x,y2,linestyle="dashed",label="cos")
plt.plot(x,y3,linestyle="dotted",label="tan")
plt.ylim([-1.5,1.5])
plt.xlim([-5,5])
plt.plot([-5,5],[0,0])
plt.plot([0,0],[-1.5,1.5])
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos & tan")
plt.legend()
plt.show()
```
